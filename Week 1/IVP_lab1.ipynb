```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
import time
```

# Question 1: RGB - Grayscale conversion


### Read images

```python
img_rose = cv2.imread('rose.jpg', 1)
plt.imshow(img_rose)
plt.show()

img_lenna = cv2.imread('lenna.tif', 1)
plt.imshow(img_lenna)
plt.show()
```

## Function using For loops

```python
def bgr2gray_for(image):
    #converts input color image to grayscale using for loops
    rows, cols = image.shape[0:2]
    b = image[:,:,0]
    g = image[:,:,1]
    r = image[:,:,2]
    
    gray = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            gray[i,j] = b[i,j]*0.2989 + g[i,j]*0.5870 + r[i,j]*0.114
                    
    return gray
```

## Function using List comprehension loops

```python
def bgr2gray_list(image):
    #converts input color image to grayscale using list comprehencsion for nested for-loops
    rows, cols = image.shape[0:2]
    b = image[:,:,0]
    g = image[:,:,1]
    r = image[:,:,2]
    
    gray = [[b[i,j]*0.2989 + g[i,j]*0.5870 + r[i,j]*0.114 for j in range(cols)] for i in range(rows)]
    return gray
```

### Generate and Plot grayscale images for rose.jpg

```python
#for rose image, convert and plot images using all 3 methods

#using list comprehension
t0 = time.time()
gray1= bgr2gray_for(img_rose)
t1 = time.time()
time_for = t1- t0

#using for loops
t0 = time.time()
gray2 = bgr2gray_list(img_rose)
t1 = time.time()
time_list = t1- t0

#using opencv library function
t0 = time.time()
cvgray = cv2.cvtColor(img_rose, cv2.COLOR_BGR2GRAY)
t1 = time.time()
time_cv = t1 - t0

#plot all images
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
ax1.imshow(img_rose)
ax1.set_title('Original')
ax2.imshow(gray1, cmap='gray')
ax2.set_title('For loops')
ax3.imshow(gray2, cmap='gray')
ax3.set_title('List comprehension')
ax4.imshow(cvgray, cmap='gray')
ax4.set_title('Opencv')
f.set_figheight(4)
f.set_figwidth(12)
plt.show()

```

### Plot processing time for rose.jpg

```python
#Plot bar graphs for time used
x = ['For loops', 'List Comprehension', 'OpenCV']
times = [time_for, time_list, time_cv ]
x_pos = [i for i in range(len(x))]
plt.bar(x_pos, times, color='blue')
plt.xlabel("<-------------Method-----------> ")
plt.ylabel("<----------Processing Time-------------> ")
plt.title(" Processing Time Vs Method ")
plt.xticks(x_pos, x)
plt.show()


print('Time in for loops:', time_for, 
      '\nTime in list comprehension:', time_list, 
      '\nTime in opencv function:', time_cv)
```

## Question 2
Q 2.Write python code  from scratch to convert Grayscale  image to Binary images using a def function and list of comprehension method and  plot processing time comparison with OpenCV and forloop methods

```python
# List comprehension function
def lc_method_q2(img):
    prop = img.shape
    img = np.asarray([[0 if img[i][j]<=127 else 255 for j in range(prop[1])] for i in range(prop[0])])
    return img

# For loop function
def for_loop_method_q2(img):
    prop = img.shape
    for i in range(prop[0]):
        for j in range( prop[1]):
            if img[i][j] <= 127:
                img[i][j] = 0
            else:
                img[i][j] = 255
    return img

# OpenCV function
def cv2_method_q2(img):
    (thresh, bin_img) = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
    return bin_img
```

### Read grayscale image

```python
messi_img = cv2.imread('messi5.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(messi_img,cmap = 'gray')
plt.title('Grayscale Image')
plt.show()
```

### For loop method

```python
# For loop method
t0 = time.time() 
binary_img_forloop = for_loop_method_q2(messi_img)
t1 = time.time()
for_loop_time = t1 - t0
plt.imshow(binary_img_forloop,cmap = 'gray')
plt.title('Binary Image: For Loop')
plt.show()
print('Time taken for for loop method is: ', for_loop_time, 'sec')
```

### List comprehension method

```python
# List comprehension method
t0 = time.time() 
binary_img_lc = lc_method_q2(messi_img)
t1 = time.time()
lc_time = t1 - t0
plt.imshow(binary_img_lc,cmap = 'gray')
plt.title('Binary Image: List Comprehension')
plt.show()
print('Time taken for list comprehension method is: ', lc_time, 'sec')
```

### OpenCV method

```python
# OpenCV method
t0 = time.time() 
binary_img = cv2_method_q2(messi_img)
t1 = time.time()
cv2_time = t1 - t0
plt.imshow(binary_img,cmap = 'gray')
plt.title('Binary Image: OpenCV')
plt.show()

print('Time taken for for OpenCV method is: ', cv2_time, 'sec')
```

### Processing time comparision

```python
x = ['For Loop', 'List Comprehension', 'OpenCV']
times = [for_loop_time, lc_time, cv2_time]
x_pos = [i for i in range(len(x))]
plt.bar(x_pos, times, color='blue')
plt.xlabel("<-------------Method-----------> ")
plt.ylabel("<----------Processing Time-------------> ")
plt.title(" Processing Time Vs Method ")
plt.xticks(x_pos, x)
plt.show()

print("Time taken for For Loop Method is :",for_loop_time)
print("Time taken for List Comprehension Method is :",lc_time)
print("Time taken for CV2 Library Method is :",cv2_time)
```
